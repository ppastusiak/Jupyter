```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,2*np.pi,100)
fig,ax=plt.subplots()
plt.style.use('default')
fig.set_size_inches(12,6)
plt.xlim(0,2*np.pi)
ax.grid()
ax.plot(x,np.sin(x))
```

```python
fig,ax=plt.subplots()
plt.style.use('default')
fig.set_size_inches(12,6)
plt.xlim(0,2*np.pi)
plt.ylim(-1.5,1.5)
ax.grid(which='both',color='#FFdddd')
ax.minorticks_on()
ax.plot(x,np.sin(x))
```

```python
fig,ax=plt.subplots()
plt.style.use('default')
fig.set_size_inches(12,6)
plt.xlim(0,2*np.pi)
plt.ylim(-1.5,1.5)
ax.axvline(x=np.pi/2,color='g',lw=0.4,label='$\pi/2$')
ax.axvline(x=np.pi/4,color='g', lw=0.4,label='$\pi/4$')
ax.axhline(y=0,color='b', lw=0.4)
ax.grid(which='both',color='#FFdddd')
ax.minorticks_on()
plt.legend()
ax.plot(x,np.sin(x))
```
