```python
import matplotlib.pyplot as plt
import numpy as np

def U4(x,y):
    #return -100/np.sqrt(x**2+y**2)
    #return (-1/np.sqrt((x-0.05)**2+y**2))+(1/np.sqrt((x+0.05)**2+y**2))
    return (-1/np.sqrt((x-0.05)**2+y**2))+(-1/np.sqrt((x+0.05)**2+y**2))+(1/np.sqrt((x)**2+(y-0.05)**2))+\
    (1/np.sqrt((x)**2+(y+0.05)**2))


x=np.linspace(-0.1,0.1,24)
y=np.linspace(-0.1,0.1,25)

X,Y = np.meshgrid(x,y)
Z=U4(X,Y)
g=np.gradient(Z)
norms = np.linalg.norm(g,axis=0)
g = np.where(norms==0,0,g/norms)
v,u=g 
v=-v
u=-u
plt.contourf(X, Y, Z,100,cmap="Greys")
plt.quiver(X,Y,u,v,color='g',lw=0.1)
```
