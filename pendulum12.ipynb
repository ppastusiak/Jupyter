```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
```

```python
def dSdt(x, S):
    theta, v = S
    return [v,(-9.81/0.25)*np.sin(theta)]
    
theta_0 = 0.0001
v_0 = 0
S_0 = (theta_0, v_0)
```

```python
t = np.linspace(0, 5, 1000)
sol = odeint(dSdt, y0=S_0, t=t, tfirst=True)
theta_sol = sol.T[0]
v_sol = sol.T[1]
```

```python
plt.plot(t,theta_sol)
plt.plot(t,v_sol)
plt.xlim(0,5)
plt.hlines(y=0,xmin=0,xmax=5)
plt.grid()
```

```python
T=[]
for i in range(0,len(v_sol)-1):
    if i>=1 and v_sol[i]*v_sol[i-1]<=0:
        T.append(t[i])

2*np.mean(np.diff(T))

    
```
