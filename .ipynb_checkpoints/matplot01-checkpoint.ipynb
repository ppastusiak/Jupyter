```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-white')

x=np.linspace(0,15,30)
y=np.sin(x)+0.1*np.random.randn(len(x))
x1=np.linspace(0,15,100)
y1=np.sin(x1)
```

```python
plt.figure(figsize=(16,9))
plt.grid()
plt.plot(x,y,'o-',linewidth=.8,color='#2596be',markersize=3)
```

```python
plt.figure(figsize=(16,9))
plt.grid()
plt.plot(x,y,'o-',linewidth=.8,color='#2596be',markersize=3)
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.figure(figsize=(16,9))
plt.grid()
plt.plot(x,y,'o',linewidth=.8,color='red',markersize=3,label='Measurments of voltage')
plt.plot(x1,y1,color='green',label='Measurments of current')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left')
plt.title('Measurments of voltage')
```
