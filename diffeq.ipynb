```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
```

```python
def dvdt(t,v):
    return 9.81-v
#funkcja definiujÄ…ca ODE
```

```python
v0=0 #warunek brzegowy

t=np.linspace(0,10,1000)
solution = odeint(dvdt,y0=[v0],t=t,tfirst=True)
```

```python
#solution.transpose()[0]
```

```python
plt.plot(t,solution[:,0])
plt.xlim(0,10)
plt.ylim(0,12)
plt.grid(True)
plt.xlabel('t [s]')
plt.ylabel('v [m/s]')
```

```python
def dSdt(x,S):
    x,v=S
    return [v,9.81-0.5*v**2]
x0=0
y0=0
S0=[x0,y0]
t=np.linspace(0,4,1000)
sol=odeint(dSdt,y0=S0,t=t,tfirst=True)
xsol=sol.transpose()[0]
vsol=sol.transpose()[1]
```

```python
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('x', color=color)
ax1.plot(t, 1000-xsol, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('v', color=color)  # we already handled the x-label with ax1
ax2.plot(t, vsol, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
