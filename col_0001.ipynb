```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y=np.sin(x)
y1=np.cos(x)
plt.figure(figsize=(10, 5))
plt.axvline(x=np.pi/3,color='red',lw=1)
plt.text(np.pi/3,-1.2,r'$\pi$/3',rotation=90)
plt.axvline(x=np.pi/4)
plt.axvline(x=np.pi/6)
plt.plot(x,y)
plt.plot(x,y1)
plt.xlim(0,2*np.pi)
plt.ylim(-1,1)
plt.grid()
plt.tight_layout()
```
