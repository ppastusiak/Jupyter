```python
from matplotlib import pyplot as plt
import numpy as np


def sinplot():
    """Example plot we'll use throughout."""
    fig, ax = plt.subplots()
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        ax.plot(x, np.sin(x + i * .5) * (7 - i))
    return ax
```

```python
ax = sinplot()
```
